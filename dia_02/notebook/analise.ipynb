```python
#ler arquivo csv
import pandas as pd 

ordens_servico_sujas = pd.read_csv('../data/ordens_servico_sujas.csv')
display(ordens_servico_sujas)
```

```python
#limpar a coluna 'status'

ordens_servico_sujas['status'] = ordens_servico_sujas['status'].str.strip().str.lower().str.capitalize().fillna('Pendente')

display(ordens_servico_sujas)
```

```python
#limpando valores ausentes

ordens_servico_sujas['valor'] = ordens_servico_sujas['valor'].fillna(0)

display(ordens_servico_sujas)
```

```python
#transformando datas str em datetime 

ordens_servico_sujas['data_os'] = pd.to_datetime(ordens_servico_sujas['data_os'])

ordens_servico_sujas.info()

display(ordens_servico_sujas)
```

```python
#nomeando nova tabela tratada
ordens_tratadas = ordens_servico_sujas

display(ordens_tratadas)
```

```python
#filtrar para apenas ordens finalizadas

ordens_tratadas = ordens_tratadas[ordens_tratadas['status'] == 'Finalizado']

display(ordens_tratadas)
```

```python
#total faturado

total_faturado = ordens_tratadas['valor'].sum()

display(f'o valor total faturado foi de R${total_faturado:.2f}')
```

```python
#quantidade de ordens finalizadas

quantidade_os = ordens_tratadas.shape[0]

display(f'foram finalizados {quantidade_os} pedidos')
```

```python
#cliente com maior faturamento
faturamento_cliente = ordens_tratadas.groupby('cliente')['valor'].sum()

display(faturamento_cliente)

top_cliente = faturamento_cliente.idxmax()
top_faturamento_cliente = faturamento_cliente.max()

display(f'{top_cliente} foi o melhor cliente do periodo, gastando no total R${top_faturamento_cliente:.2f}')
```

```python
#serviço com maior valor médio

media_servico = ordens_tratadas.groupby('servico')['valor'].mean()

display(media_servico)

top_servico = media_servico.idxmax()
top_media_servico = media_servico.max()

display(f'{top_servico} trouxe a maior media de faturamento: R${top_media_servico:.2f}')
```

```python
#primeiro e último dia com ordem finalizada

primeira_data = ordens_tratadas['data_os'].min()
ultima_data = ordens_tratadas['data_os'].max()

display(f'o primeiro pedido foi registrado em {primeira_data} e o ultimo em {ultima_data}')
```

```python
# criar relatorio

relatorio = pd.DataFrame({
    'metrica': [
        'total_faturado',
        'qtd_os_finalizadas',
        'cliente_maior_faturamento',
        'valor_cliente_maior_faturamento',
        'servico_maior_media',
        'valor_servico_maior_media',
        'primeira_data_finalizada',
        'ultima_data_finalizada'
    ],
    'valor': [
        total_faturado,
        quantidade_os,
        top_cliente,
        top_faturamento_cliente,
        top_servico,
        top_media_servico,
        primeira_data,
        ultima_data
    ]
}
)

display(relatorio)
```

```python
# exportando relatorio
relatorio.to_csv('../output/relatorio.csv', index=False)
```
