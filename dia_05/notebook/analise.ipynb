```python
# Objetivo

# Qual foi o faturamento total?

# Quanto cada cliente faturou?

# Quantas ordens cada cliente realizou?

# Qual percentual do faturamento total cada cliente representa?

# Qual cliente tem maior participação no faturamento?
```

```python
# criando df

import pandas as pd 

df = pd.read_csv('../data/clientes_faturamento.csv')
display(df.head(3))
```

```python
# Qual foi o faturamento total?
faturamento_total = df['valor'].sum()
display(f'Faturamento total: R${faturamento_total:.2f}')
```

```python
# Quanto cada cliente faturou?

faturamento_cliente = df.groupby('cliente')['valor'].sum()

for cliente, valor in faturamento_cliente.items():
    display(f'{cliente}: R$ {valor:.2f}')
```

```python
# Quantas ordens cada cliente realizou?

ordens_cliente = df.groupby('cliente')['valor'].count()

for cliente, pedidos in ordens_cliente.items():
    display(f'{cliente} teve {pedidos} pedidos')
```

```python
# Qual percentual do faturamento total cada cliente representa?

porcentagem_cliente = []
for cliente, valor in faturamento_cliente.items():
    
    display(f'{cliente} representa {valor/faturamento_total*100:.1f}% do faturamento total')

    dados_cliente={'cliente': cliente, '%_faturamento': valor/faturamento_total*100}
    porcentagem_cliente.append(dados_cliente)

porcentagem_cliente = pd.DataFrame(porcentagem_cliente).sort_values('%_faturamento')
porcentagem_cliente['%_faturamento'] = porcentagem_cliente['%_faturamento'].round(1)
porcentagem_cliente.to_csv('../output/porcentagem_cliente.csv', index=False)

display(porcentagem_cliente)
```

```python
# melhor cliente

melhor_cliente = df.groupby('cliente')['valor'].sum().idxmax()

display(f'{melhor_cliente} é o melhor cliente')
```

```python
# criando relatorio

relatorio = pd.DataFrame({
    'faturamento_total': faturamento_cliente,
    'qtd_ordens': ordens_cliente,
    'participacao_percentual': faturamento_cliente/faturamento_total*100
})

relatorio['participacao_percentual'] = (
    relatorio['participacao_percentual']
    .round(1)
)

relatorio = (
    relatorio
    .sort_values(
        by='participacao_percentual',
        ascending=False
    )
)

display(relatorio)
```

```python
# exportando relatorio
relatorio.to_csv('../output/relatorio_clientes.csv')
```
