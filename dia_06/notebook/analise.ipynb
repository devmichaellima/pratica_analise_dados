```python
# criando dataframe

import pandas as pd 

df = pd.read_csv('../data/servicos_faturamento.csv')

display(df.head(5))
```

```python
# filtrando por status

df = df[df['status']== 'Finalizado']
display(df.info())
```

```python
# Qual foi o faturamento total considerando apenas ordens finalizadas?

faturamento_total = df['valor'].sum()

display(f'faturamento total: R${faturamento_total}')
```

```python
# Quanto cada serviço faturou?

faturamento_servico = df.groupby('servico')['valor'].sum()
display(faturamento_servico)
```

```python
# Quantas ordens finalizadas cada serviço teve?

ordens_finalizadas = df.groupby('servico')['id_os'].count()
display(ordens_finalizadas)
```

```python
# Qual foi o ticket médio de cada serviço?

media_servico = df.groupby('servico')['valor'].mean().round(1)
display(media_servico)
```

```python
# Qual percentual do faturamento total cada serviço representa?

porcem_faturamento = (faturamento_servico/faturamento_total*100).round(1)
display(porcem_faturamento)
```

```python
# Qual serviço teve maior faturamento?
top_servico = faturamento_servico.idxmax()
display(top_servico)
```

```python
# Qual serviço teve maior ticket médio?
maior_media_servico = media_servico.max()
display(maior_media_servico)
```

```python
# criando relatorio
relatorio = pd.DataFrame({
    'faturamento_total': faturamento_servico,
    'qtd_ordens': ordens_finalizadas,
    'ticket_medio': media_servico,
    'participacao_percentual': porcem_faturamento
}).reset_index()

display(relatorio)
```

```python
# exportando relatorio
relatorio.to_csv('../output/relatorio.csv')
```
