```python
#objetivo

# - Qual foi o faturamento total?
# - Qual dia teve maior faturamento?
# - Qual dia teve menPor faturamento?
# - Qual foi o faturamento médio por dia?
# - Quantos dias tiveram vendas registradas?

# passo a passo

# ler arquivo csv
# limpeza e tratamento
# analise
```

```python
# lendo arquivo
import pandas as pd 

vendas_periodo = pd.read_csv('../data/vendas_periodo.csv')

display(vendas_periodo)
```

```python
# limpeza e tratamento
vendas_periodo['data_os'] = pd.to_datetime(vendas_periodo['data_os'])

vendas_periodo.info()
display(vendas_periodo)
```

```python
# - Qual foi o faturamento total?

faturamento_total = vendas_periodo['valor'].sum()

display(f'o faturamento total foi de R${faturamento_total:.2f}')
```

```python
# - Qual dia teve maior faturamento?
faturamento_dia = vendas_periodo.groupby('data_os')['valor'].sum()
melhor_dia = faturamento_dia.idxmax()
faturamento_melhor_dia = faturamento_dia.max()

display(f'{melhor_dia} foi o dia com maior faturamento: R${faturamento_melhor_dia:.2f}')
```

```python
# - Qual dia teve menPor faturamento?

pior_dia = faturamento_dia.idxmin()
faturamento_pior_dia = faturamento_dia.min()

display(f'{pior_dia} foi o dia com menor faturamento: R${faturamento_pior_dia:.2f}')
```

```python
# - Qual foi o faturamento médio por dia?
media_dia = faturamento_dia.mean()

display(f'{media_dia:.2f} é o ticket medio por dia')
```

```python
# Quantos dias tiveram vendas registradas?
contagem_dias_registrado = faturamento_dia.count()

display(f'foram registrado {contagem_dias_registrado} dias com pedidos')
```

```python
# criando relatorio

relatorio = pd.DataFrame({
    'metrica':[
        'total_faturado',
        'dia_maior_faturamento',
        'valor_dia_maior_faturamento',
        'dia_menor_faturamento',
        'valor_dia_menor_faturamento',
        'faturamento_medio_diario',
        'dias_com_vendas'
    ],
    'valor':[
        faturamento_total,
        melhor_dia,
        faturamento_melhor_dia,
        pior_dia,
        faturamento_pior_dia,
        media_dia,
        contagem_dias_registrado
    ]

})
```

```python
# exportando relatorio csv

relatorio.to_csv('../output/relatorio.csv', index=False)
```
