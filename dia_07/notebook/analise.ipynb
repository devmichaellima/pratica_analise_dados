```python
# A empresa precisa responder:

# Qual foi o faturamento total considerando apenas ordens finalizadas?

# Quanto cada cliente faturou?

# Quantas ordens finalizadas cada cliente realizou?

# Qual foi o ticket médio por cliente?

# Quanto cada cidade faturou?

# Qual cidade teve maior faturamento?
```

```python
# criando dataframes
import pandas as pd

clientes = pd.read_csv('../data/clientes.csv')
ordens = pd.read_csv('../data/ordens_servico.csv')

display(clientes, ordens)
```

```python
# unindo tabelas

df = clientes.merge(ordens, how='right', on='id_cliente')
display(df.head(4))
```

```python
# filtrando por status finalizado
df = df[df['status']=='Finalizado']
display(df.head(4))
```

```python
# Qual foi o faturamento total considerando apenas ordens finalizadas?
faturamento_total = df['valor'].sum()
display(faturamento_total)
```

```python
# Quanto cada cliente faturou?
faturamento_cliente = df.groupby('nome_cliente')['valor'].sum()
display(faturamento_cliente)
```

```python
# Quantas ordens finalizadas cada cliente realizou?
ordens_cliente = df.groupby('nome_cliente')['id_os'].count()
display(ordens_cliente)
```

```python
# Qual foi o ticket médio por cliente?
media_cliente = df.groupby('nome_cliente')['valor'].mean().round(1)
display(media_cliente)
```

```python
# Quanto cada cidade faturou?
faturamento_cidade = df.groupby('cidade')['valor'].sum()
display(faturamento_cidade)
```

```python
# Qual cidade teve maior faturamento?
top_cidade = faturamento_cidade.idxmax()
display(top_cidade)
```

```python
cidade_cliente = df.groupby('nome_cliente')['cidade'].first()
display(cidade_cliente)
```

```python
id_cliente = df.groupby('nome_cliente')['id_cliente'].first()
display(id_cliente)
```

```python
# criando relatorio
relatorio = pd.DataFrame({
    'id_cliente': id_cliente,
    'cidade': cidade_cliente,
    'faturamento_total': faturamento_cliente,
    'qtd_ordens': ordens_cliente,
    'ticket_medio': media_cliente
}).reset_index().sort_values('id_cliente')

display(relatorio)
```

```python
# exportando relatorio
relatorio.to_csv('../output/relatorio.csv')
```
