```python
# Objetivo

# - Qual foi o faturamento total da empresa?
# - Quanto cada cliente faturou?
# - Quem foi o cliente que mais gerou receita?
# - Qual foi o faturamento médio por cliente?
# - Quais são os 3 clientes com maior faturamento?

# Etapas

# - extrair dados
# - transformar dados
# - analisar os dados
```

```python
# Lendo arquivo
import pandas as pd

df = pd.read_csv('../data/clientes_faturamento.csv')
display(df)
```

```python
# - Qual foi o faturamento total da empresa?

faturamento_total = df['valor'].sum()
display(f'total faturado: R${faturamento_total:.2f}')
```

```python
# - Quanto cada cliente faturou?

faturamento_cliente = df.groupby('cliente')['valor'].sum()
# display(faturamento_cliente)

for cliente, valor in faturamento_cliente.items():
    display(f'{cliente} faturou: R${valor:.2f}')
```

```python
# - Quem foi o cliente que mais gerou receita?

melhor_cliente = faturamento_cliente.idxmax()
display(f'{melhor_cliente} foi o cliente com maior faturamento')
```

```python
# - Qual foi o faturamento médio por cliente?

media_cliente = faturamento_cliente.mean()
display(f'{media_cliente:.2f} foi a media faturada por cliente')
```

```python
# - Quais são os 3 clientes com maior faturamento?

top_clientes = faturamento_cliente.sort_values(ascending=False).head(3)

for posicao,(cliente, valor) in enumerate(top_clientes.items(), start=1):
    display(f'{posicao}º colocado: {cliente} com R${valor:.2f}')
```

```python
# criando relatorio

relatorio = pd.DataFrame({
    'metrica':[
        'faturamento_total',
        'cliente_maior_faturamento',
        'valor_maior_faturamento',
        'faturamento_medio_cliente'
    ],
    'valor':[
        faturamento_total,
        faturamento_cliente.idxmax(),
        faturamento_cliente.max(),
        media_cliente
    ]
})
```

```python
# exportando relatorio
relatorio.to_csv('../output/relatorio.csv', index=False)
```

```python
# exportando top clientes
top_clientes.to_csv(
    '../output/top_clientes.csv',
    header=['faturamento']
)
```
